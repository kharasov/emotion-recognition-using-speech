```python
def extract_segment(x, time_segment, sr):
    """
    Extract audio 
    
    x - data
    time_segment - array or tuple with start and end of time interval
    sr - sampling rate of audio 
    """
    return x[int(time_segment[0] * sr) : int(time_segment[1] * sr)]
```
